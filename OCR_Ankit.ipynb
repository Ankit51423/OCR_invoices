```python
from pdf2image import convert_from_path
import pytesseract
import cv2
```

```python
#Here i am  Convert PDF to images.
images = convert_from_path('invoice.pdf', 500, poppler_path=r"C:\\Program Files\\poppler-24.02.0\\Library\\bin")

# here is a Path to tesseract executable.
pytesseract.pytesseract.tesseract_cmd = r"C:\\Program Files\\Tesseract-OCR\\tesseract.exe"

# here is Loop through each image and extract text.
for i in range(len(images)):
    # Save the image (optional, if you want to keep the images)
    images[i].save('page' + str(i) + '.jpg', 'JPEG')
    
# here is Read the image.
    img = cv2.imread('page' + str(i) + '.jpg')
    
# here is Display image.
    cv2.imshow('Page ' + str(i), img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# here is Extract text.
    text = pytesseract.image_to_string(img)
    print(f"Text from page {i}:\n{text}\n")
```
